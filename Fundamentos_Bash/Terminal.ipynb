## Wildcards
- Las wildcards o comodines son una serie de caracteres especiales que nos permiten encontrar patrones o realizar bsquedas ms avanzadas. Es aplicable para archivos y directorios.
- Las wildcards te sirven para realizar seccionamiento de archivos o directorios, adems de `ls` los wildcards tambin pueden usarse con cualquier comando que realice la manipulacin de archivos como `mv`, `cp`, y `rm`.
- Comandos:
  + `ls *"tipo de archivo que se desea buscar"`: muestra los tipos de archivos seleccionados.
    * Ej: `ls *.txt`
  + `ls "nombre del archivo"*`: muestra los archivos con ese nombre.
    * Ej: `ls TI*`
  + `ls "nombre del archivo"?`: esto sirve para buscar todo lo que comience por la cadena de caracteres "nombre del archivo".
  + `ls "nombre archivo"?`: sirve para buscar con 2 caracteres y as sucesivamente.
  + `ls "nombre archivo"?*`: puedes combinar wildcards.
  + `ls [[:"caracterstica especfica":]]*`: esto sirve para buscar un archivo con caractersticas especficas.
    * `ls [[:upper:]]*`: este comando busca nicamente lo que inicie con mayscula.
  + `ls ["caracter especfico"]*`.
    * `ls [ad]*`: esto muestra los archivos que inician con las letras a y d.

### Qu son las entradas y salidas de la terminal
- En la consola nosotros generamos una entrada cuando escribimos y una salida casi siempre que ejecutamos un comando
- A las entradas tpicamente se les suele llamar Standard Input y a las salidas Standard Output, adems se les suele abreviar como stdin y stdout

#### Redirigir el Standard Output
- `>` + `"nombre archivo.extension"`: sirve para enviar el stdout al archivo determinado.
- `>>` + `"nombre archivo.extension"`: sirve para concatenar (pegar dos cosas) las salidas de un comando a un archivo ya existente.
- Para redirigir un error el comando que usamos es `"el error" + 2` (el 2 se pone por el file descriptor 2, el cual es el standard error) + `"nombre archivo.txt"`.
- Para redirigir el standard output y el standard error se usa el comando: `> "nombre archivo.extension" 2>&1` (el `2>&1` se pone para que redirija el standard output 1 y el standard error 2).
- Ejemplos:
  + `ls -l > output.txt`
  + `ls -l ./Certificados >> output.txt`
  + `ls -l 2> error.txt`
  + `ls -l > output.txt 2>&1`

- `cat < error.txt`



### Pipe Operator
- El Pipe operator es un operador que permite tomar la salida de un comando y pasarla como entrada a otro comando.
- El Pipe operator es esto: `|` y se usa as:
  + `"comando" | "comando"` y as sucesivamente hasta que sea necesario.
- Comandos:
  + `echo "palabra deseada"`: sirve para generar un stdout (una palabra).
  + `cat "nombre archivo.txt" "nombre archivo.txt"`: sirve para concatenar la salida de los dos archivos.
  + `tee "nombre archivo.txt"`: esto sirve para hacer lo mismo que el `>` pero tambin muestra el output en la terminal.
  + `sort`: ordena los elementos de la entrada proporcionada.

## Encadenando Comandos: Operadores de Control
- Los operadores de control son smbolos reservados por la terminal que nos permiten encadenar comandos.
- Si usas constantemente la tecla enter para ejecutar varios comandos, puedes evitarlo usando el operador `;` que separa los comandos que estamos ejecutando.
- Comando para ejecutar comandos sincrnicamente, es decir, uno detrs de otro:
  + El comando es `;`.
  + Ej: `ls; mkdir Chamba; cal`
- Comando para ejecutar asincrnicamente, es decir, se ejecutan al mismo tiempo:
  + El comando es `&`.
  + Ej: `ls & date & cal`
- Comando de manera condicional and, es decir, ambos deben ser verdad:
  + El comando es `&&`.
  + Ej: `mkdir test && cd test`
- Comando de manera condicional or, es decir, uno debe ser verdad:
  + El comando es `||`.
  + Ej: `cd kgqwd || touch archivo.txt`
- Puedes combinar estos operadores de control para crear flujos de trabajo ms complejos en la terminal.

### Modificando permisos en la terminal
- Para cambiar los permisos de un archivo se usa el comando `chmod`:
  + `chmod "permiso del usuario" "permiso del grupo" "permiso del world" "archivo"`
    * Ej: `chmod 755 mitexto.txt`
  + `chmod "smbolo"-"permiso" "archivo"`: con este comando se puede quitar permisos.
    * Ej: `chmod u-r mitexto.txt`
  + `chmod "smbolo"+"permiso" "archivo"`: con este comando se puede poner permisos.
    * Ej: `chmod u+r mitexto.txt`
- Para gestionar usuarios se usa el comando `whoami` o `id`.
- Para cambiar de usuario se usa `su` (Switch User):
  + Se escribe `sudo su` o `su root`: esto te pone en modo admin, osea modo root (**Advertencia: el root tiene poder para hacer y deshacer con el sistema operativo, puedes eliminar cosas que no deberas eliminar y puede hacer mucho desastre**).
  + Para volver a tu usuario escribes `su` y tu usuario.

## Usando el comando grep

**Qu hace grep?**
- `grep` nos permite encontrar coincidencias de texto en archivos utilizando cadenas de texto o expresiones regulares.
  + **Ejemplo de uso bsico**: `grep "palabra a encontrar" "archivo"` o `grep "Expresin Regular" "archivo"`

**Qu hace el parmetro -i en grep?**
- El parmetro `-i` hace que la bsqueda sea insensible a maysculas y minsculas.
  + **Ejemplo**: `grep -i "palabra" "archivo"`

**Qu hace el parmetro -v?**
- El parmetro `-v` devuelve las lneas que no contienen la coincidencia.
  + **Ejemplo**: `grep -v "palabra" "archivo"`

**Qu hace -c?**
- El parmetro `-c` cuenta el nmero de lneas que contienen la coincidencia.
  + **Ejemplo**: `grep -c "palabra" "archivo"`

**Qu hace -m?**
- El parmetro `-m` limita el nmero de coincidencias devueltas a un nmero especfico de lneas.
  + **Ejemplo**: `grep -m 5 "palabra" "archivo"`

**Qu hace wc?**
- `wc` es un comando que cuenta la cantidad de palabras, lneas y bytes en un archivo.
  + **Parmetros de wc**:
    * `-l`: cuenta el nmero de lneas.
    * `-w`: cuenta el nmero de palabras.
    * `-c`: cuenta el nmero de bytes.
  + **Ejemplo de uso**: `wc -l "archivo"`

**Combinar opciones de grep**
- Puedes combinar varias opciones de `grep` para realizar bsquedas ms especficas.
  + **Ejemplos de combinacin**:
    * `grep -iv "palabra" "archivo"`: busca ignorando maysculas y muestra lneas sin la coincidencia.
    * `grep -ci "palabra" "archivo"`: cuenta las lneas que contienen la coincidencia, ignorando maysculas.
    * `grep -vi "palabra" "archivo"`: muestra lneas sin la coincidencia, ignorando maysculas.
    * `grep -vm 5 "palabra" "archivo"`: muestra hasta 5 lneas que no contienen la coincidencia.

## Comprimiendo y Descomprimiendo Archivos TAR y ZIP

### Comprimiendo archivos TAR
- `tar -c` seguido del nombre del archivo `.tar` crea un nuevo archivo tar.
- `tar -cf "nombre archivo".tar`: indica que se est comprimiendo un archivo y se especifica el nombre.
- `tar -cvf "nombre archivo".tar`: muestra en detalle (verbose) lo que se est comprimiendo.
- `tar -cvf "nombre".tar "archivos"`: combina las opciones de compresin, verbosidad y archivo.
- `tar -czvf "nombre archivo".tar.gz`: comprime y utiliza gzip para manejar archivos ms grandes.
- `tar -czvf "nombre".tar.gz "archivos"`: detalla la operacin de compresin con gzip, verbosidad y archivo especificado.

### Descomprimiendo archivos TAR
- `tar -x "nombre archivo".tar`: descomprime un archivo tar.
- `tar -xf "nombre archivo".tar`: indica que se est descomprimiendo un archivo y se especifica el nombre.
- `tar -xvf "nombre archivo".tar`: muestra en detalle (verbose) lo que se est descomprimiendo.
- `tar -xvf "nombre".tar "archivos"`: combina las opciones de descompresin, verbosidad y archivo.
- `tar -xzvf "nombre archivo".tar.gz`: descomprime y maneja archivos comprimidos con gzip.
- `tar -xzvf "nombre".tar.gz "archivos"`: detalla la operacin de descompresin con gzip, verbosidad y archivo especificado.

### Comprimiendo archivos ZIP
- `zip "nombre archivo".zip`: comprime archivos individuales en un archivo zip.
- `zip -r "nombre archivo".zip "carpeta"`: comprime una carpeta completa, incluyendo subcarpetas y archivos.

### Descomprimiendo archivos ZIP
- `unzip "nombre archivo".zip`: descomprime un archivo zip.

## Manejo de Procesos
### Visualizacin de Procesos
- **`ps`**: Este comando se usa para ver los procesos activos en la terminal. Puedes usar diferentes opciones para ver ms detalles:
  + `ps aux`: Muestra todos los procesos en ejecucin con detalles completos.
  + `ps -e`: Muestra todos los procesos del sistema.
  + `ps -u usuario`: Muestra los procesos de un usuario especfico.

- **`top`**: Este comando proporciona una vista dinmica en tiempo real de los procesos en ejecucin del sistema. Es til para monitorizar la carga del sistema y los recursos utilizados por los procesos:
  + Dentro de `top`, puedes usar teclas como `k` para matar procesos, `u` para filtrar por usuario, y `P` para ordenar por uso de CPU.

- **`jobs`**: Este comando proporciona una vista dinmica en tiempo real de los procesos en ejecucin del sistema en backgound.
### Gestin de Procesos
- **`kill`**: Este comando se utiliza para enviar seales a procesos. La ms comn es la seal de terminacin (`SIGTERM`) para pedir a un proceso que se detenga de manera segura. Si eso falla, se puede usar `SIGKILL` para forzar el cierre:
  + `kill 1234`: Enva la seal `SIGTERM` al proceso con PID 1234.
  + `kill -9 1234`: Enva la seal `SIGKILL` al proceso con PID 1234, forzando su terminacin.
  + `kill -l`: Lista todas las seales que se pueden enviar con `kill`.

- **`htop`** (si est instalado): Una versin mejorada de `top` con una interfaz ms amigable y opciones adicionales para gestionar procesos de forma interactiva.
