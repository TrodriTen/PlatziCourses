# Fundamentos en ingenieria software

## Byte y bits
- **Bit**
  + Es el codigo binario de 0 y 1
- **Byte**
  + Unidad unificada de memoria
  + Tiene 8 espacios o bits
  + El sistema de 1 y 0 esta determinado por potencias de `2^x`
  + Los caracteres o letras se le asignan numeros que eson equivalentes a un byte(`UNICODE`, `ASCII`) para que el computador entienda cual letra es por el sistema binario
  + Assembler son bytes especiales del procesador, es considerado el primer codigo de la historia computacional
  + Byte < 256
  + Estandar *UTF-8 y UTF-16*
  + Los emojis son bytes dobles, es decir, estan en el estandar UTF-16
  + Funcionamiento del computador en eventos:
    * 
  + `API` - Application Program Interface: Programna utilizado por programadores para comunicarse entre si sin hablar directamente
  + `REST` - Representational State Transfer: Protocolo establecido para el envio de datos
  + `AJAX` - Asynchonous JavaScript and XML: API la cual se pueden enviar datos sin recargar el sitio web
  + `JSON` - JavaScript Objetc Notation: Formato de almacenamiento y transporte de datos
  +  `SSH` - Secure Shell
  + **Protocolos**:
    *  `HTTPS` - Hypertext Transfer Protocol Secure: Protocolo de transferencia de datos de forma segura
    *  `FTP` - File Transfer Protocol: Enviar archivos grandes (Obsoleto)
    *  `URL` - Unifor Resource Locator: Conforma la direccion de los sitios web
    *  `DNS` - Domain Name System: Son una serie de servidores globales, es decir una DB gigantesca
    *  `IP` - Internet Protocol: Va de la mano con la URL
    *  `SMTP` - Simple Mail Trasnfer Protocol: Es un protocolo de envio para que los servidores de correo transmitan los mails
    *  `UDP` - User Datagram Protocol
    *  `POP3` - Post Office Protocol: Recibir correos desde un servidor remoto

## Procesadores y arquitectura de la CPU
- CPU - Central Processing Unit:
  * Es quien interpreta las instrucciones de un programa mediante operaciones aritmeticas , logicas y de entrada/salida del sistema
  * `Hertz (Hz)`: es una unidad de frecuencia, esta representa los ciclos por segundo, en CPU es la activacion por segundo.
  * `GigaHertz (GHz)`: Es un multiplo de Hertz, equivale a 10Hz
  * `Ciclo`: Repeticion de un suceso en el tiempo
  * `Cores`: Numero de CPUs en un mismo chip, los cores determinan cuantas actividades se pueden hacer en paralelo antes de que haya concurrencia.
  * `BIOS`: Es un software (firware) que inicia y comprueba todos los componentes de hardware, es un sistema de operativo de arranque.
  * `Hard Disk`: Dispositivo donde se almacena toda la informacion que se procesa en la computadora incluyendo el OS.
  * `RAM`: Memoria principal, es donde se ejecuta la gran mayoria del software (OS, programas, etc)
